```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from xgboost import XGBClassifier
from sklearn.preprocessing import StandardScaler
```

```python
# Load data from Excel file
df = pd.read_excel('F:\Image CHD\ImageCHD_dataset\Training Data\glcm_features.xlsx')
```

```python
# Extract features and target variable
X = df[['Contrast', 'Dissimilarity', 'Homogeneity', 'Energy', 'Correlation']].values
y = df['ASD'].values
```

```python
# Standardize features by removing the mean and scaling to unit variance
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Create XGBoost classifier
xgb_classifier = XGBClassifier(random_state=42)
```

```python
# Train the XGBoost classifier
xgb_classifier.fit(X_scaled, y)
```

```python
# Predict the target variable
predicted = xgb_classifier.predict(X_scaled)
```

```python
# Calculate evaluation metrics
accuracy = accuracy_score(y, predicted)
precision = precision_score(y, predicted)
recall = recall_score(y, predicted)
f1 = f1_score(y, predicted)
```

```python
# Print evaluation metrics
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
